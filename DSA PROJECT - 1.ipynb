```python
from mpl_toolkits.basemap import Basemap
```

```python
plt.rcParams["figure.figsize"]=13,13
m=Basemap()
m.drawcoastlines()
 
plt.show();
```

```python
lats = df['latitude'].values
lons = df['longitude'].values
magnitudes = df['mag'].values
```

```python
# Scatter plot earthquake locations on the map
x, y = m(lons, lats)
m.scatter(x, y, c=magnitudes, s=10, cmap='Reds', alpha=0.7, edgecolors='k', linewidth=0.5)
plt.colorbar(label='Magnitude')
plt.title('Spatial Distribution of Earthquakes')
plt.show()
```

```python
plt.rcParams["figure.figsize"] = 13, 13
m = Basemap(projection='mill', llcrnrlat=-60, urcrnrlat=90, llcrnrlon=-180, urcrnrlon=180, resolution='c')
m.drawcoastlines()
m.drawcountries()
# Scatter plot earthquake locations on the map
x, y = m(lons, lats)
m.scatter(x, y, c=magnitudes, s=10, cmap='Reds', alpha=0.7, edgecolors='k', linewidth=0.5)
plt.colorbar(label='Magnitude')
plt.title('Spatial Distribution of Earthquakes')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['mag'], bins=30, kde=True, color='blue')
plt.title('Distribution of Magnitudes')
plt.xlabel('Magnitude')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['depth'], bins=30, kde=True, color='blue')
plt.xlabel('depth')
plt.show()
```

```python
dist_num_data=df.select_dtypes(exclude="object").drop(['latitude','longitude','depth','mag'],axis=1)
fig,axs=plt.subplots(len(dist_num_data.columns)//2,2,figsize=(12,14))
axs=axs.flatten()
for ax,data in zip(axs,dist_num_data):
    sns.histplot(ax=ax,data=dist_num_data[data],kde=True)
    ax.set_title("{} Distribution".format(str(dist_num_data[data].name).capitalize()))
    
fig.tight_layout(pad=.9)
```

```python
 plt.figure(figsize=(12, 8))
 sns.scatterplot(x='longitude', y='latitude', hue='mag', size='mag',sizes=(20, 300),
                data=df, palette='viridis', alpha=0.7)
 plt.title('Earthquake Locations')
 plt.xlabel('Longitude')
 plt.ylabel('Latitude')
 plt.legend(title='Magnitude')
 plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.scatterplot( x=df["depth"], y=df["mag"], hue=df["mag"], palette="viridis", alpha=0.6)
plt.title('Depth vs Magnitude of Earthquakes')
plt.xlabel('Depth (km)')
plt.ylabel('Magnitude')
plt.show()
```

```python
df['time'] = pd.to_datetime(df['time'], yearfirst=True)
df['month'] = df['time'].dt.month_name()
df['year'] = df['time'].dt.year
df['day'] = df['time'].dt.day
df['date'] = df['time'].dt.date
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(df.groupby('date')['day'].count().reset_index(name='Count'), x='date', y='Count')
plt.grid(axis='y',linewidth=0.1)
plt.title('number of earthquake occurrences')
plt.show()
```

```python
sns.barplot(df.groupby([
    'month'])['day'].count().sort_values(ascending=False).sort_values(ascending=False).reset_index(name='Count'), x='month', y='Count')
plt.xticks(rotation=45) 
plt.title('earthquake occurrences by month')
plt.grid(axis='y', color='black', linestyle='-', linewidth=0.1)
plt.show()
```
